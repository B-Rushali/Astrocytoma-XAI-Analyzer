```python
# Custom PyTorch Dataset class for loading images and masks from a DataFrame.
class BrainDataset(data.Dataset):
    def __init__(self, df, transform=None):
        self.df = df
        self.transform = transform

    def __len__(self):
        return len(self.df)

    def __getitem__(self, idx):
        image = cv2.imread(self.df.iloc[idx, 0])
        image = np.array(image)/255.
        mask = cv2.imread(self.df.iloc[idx, 1], 0)
        mask = np.array(mask)/255.

        if self.transform is not None:
            aug = self.transform(image=image, mask=mask)
            image = aug['image']
            mask = aug['mask']

        image = image.transpose((2,0,1))
        image = torch.from_numpy(image).type(torch.float32)
        image = tt.Normalize((0.485, 0.456, 0.406), (0.229, 0.224, 0.225))(image)
        mask = np.expand_dims(mask, axis=-1).transpose((2,0,1))
        mask = torch.from_numpy(mask).type(torch.float32)

        return image, mask
```

```python
# Defines the DoubleConv class, consisting of two consecutive
# convolutional layers each followed by BatchNorm and ReLU.

class DoubleConv(nn.Module):
    """(convolution => [BN] => ReLU) * 2"""
    def __init__(self, in_channels, out_channels, mid_channels=None):
        super().__init__()
        if not mid_channels:
            mid_channels = out_channels
        self.double_conv = nn.Sequential(
            nn.Conv2d(in_channels, mid_channels, kernel_size=3, padding=1),
            nn.BatchNorm2d(mid_channels),
            nn.ReLU(inplace=True),
            nn.Conv2d(mid_channels, out_channels, kernel_size=3, padding=1),
            nn.BatchNorm2d(out_channels),
            nn.ReLU(inplace=True))
    def forward(self, x):
        return self.double_conv(x)


# Defines the Down class, which performs downscaling using MaxPool
# followed by DoubleConv.
class Down(nn.Module):
    """Downscaling with maxpool then double conv"""
    def __init__(self, in_channels, out_channels):
        super().__init__()
        self.maxpool_conv = nn.Sequential(
            nn.MaxPool2d(2),
            DoubleConv(in_channels, out_channels))
    def forward(self, x):
        return self.maxpool_conv(x)


# Defines the Up class, which performs upscaling followed by DoubleConv.
class Up(nn.Module):
    """Upscaling then double conv"""
    def __init__(self, in_channels, out_channels, bilinear=True):
        super().__init__()
        if bilinear:
            self.up = nn.Upsample(scale_factor=2, mode='bilinear', align_corners=True)
            self.conv = DoubleConv(in_channels, out_channels, in_channels//2)
        else:
            self.up = nn.ConvTranspose2d(in_channels, in_channels//2, kernel_size=2, stride=2)
            self.conv = DoubleConv(in_channels, out_channels)
    def forward(self, x1, x2):
        x1 = self.up(x1)
        diffY = x2.size()[2] - x1.size()[2]
        diffX = x2.size()[3] - x1.size()[3]

        x1 = F.pad(x1, [diffX//2, diffX-diffX//2,
                        diffY//2, diffY-diffY//2])
        x = torch.cat([x2, x1], dim=1)
        return self.conv(x)

# Defines the OutConv class, which performs a 1x1 convolution followed by Sigmoid activation.
class OutConv(nn.Module):
    def __init__(self, in_channels, out_channels):
        super(OutConv, self).__init__()
        self.conv = nn.Sequential(
            nn.Conv2d(in_channels, out_channels, kernel_size=1),
            nn.Sigmoid())
    def forward(self, x):
        return self.conv(x)
```

```python
import torch
import torch.nn as nn
from torchvision import models

class SqueezeNetUNet(nn.Module):
    def __init__(self, in_channels, out_channels):
        super(SqueezeNetUNet, self).__init__()

        # Load pretrained SqueezeNet model
        squeezenet = models.squeezenet1_1(pretrained=True)

        # Take the convolution layers from SqueezeNet (exclude the classifier)
        self.encoder = nn.Sequential(*list(squeezenet.features.children()))

        # Decoder part (U-Net like upsampling)
        self.upconv1 = self.upconv_block(512, 256)
        self.upconv2 = self.upconv_block(256, 128)
        self.upconv3 = self.upconv_block(128, 64)
        self.upconv4 = self.upconv_block(64, 32)

        # Final output layer (convolution to map to the required output channels)
        self.final_conv = nn.Conv2d(32, out_channels, kernel_size=1)

    def forward(self, x):
        # Encoder: pass input through SqueezeNet
        enc_out = self.encoder(x)

        # Decoder: upsample and decode features
        x = self.upconv1(enc_out)
        x = self.upconv2(x)
        x = self.upconv3(x)
        x = self.upconv4(x)

        # Final convolution to get output segmentation mask
        out = self.final_conv(x)

        return out

    def upconv_block(self, in_channels, out_channels):
        # Upsample block with convolution, batch normalization, and ReLU
        return nn.Sequential(
            nn.ConvTranspose2d(in_channels, out_channels, kernel_size=2, stride=2),
            nn.BatchNorm2d(out_channels),
            nn.ReLU(inplace=True)
        )
```

```python
# Defines the UNet class, a convolutional neural network architecture
# for semantic segmentation.
class UNet(nn.Module):
    def __init__(self, n_channels, n_classes, bilinear=True):
        super(UNet, self).__init__()
        self.n_channels = n_channels
        self.n_classes = n_classes
        self.bilinear = bilinear

    # Encoder (downsampling path)
        self.inc = DoubleConv(n_channels, 64)
        self.down1 = Down(64, 128)
        self.down2 = Down(128, 256)
        self.down3 = Down(256, 512)
        factor = 2 if bilinear else 1
        self.down4 = Down(512, 1024//factor)

    # Decoder (upsampling path)
        self.up1 = Up(1024, 512//factor, bilinear)
        self.up2 = Up(512, 256//factor, bilinear)
        self.up3 = Up(256, 128//factor, bilinear)
        self.up4 = Up(128, 64, bilinear)

    # Output layer
        self.outc = OutConv(64, n_classes)

    def forward(self, x):

    # Encoder path
        x1 = self.inc(x)
        x2 = self.down1(x1)
        x3 = self.down2(x2)
        x4 = self.down3(x3)
        x5 = self.down4(x4)

    # Decoder path with skip connections
        x = self.up1(x5, x4)
        x = self.up2(x, x3)
        x = self.up3(x, x2)
        x = self.up4(x, x1)

    # Final output
        logits = self.outc(x)
        return logits
```

```python
# Function to calculate the Dice coefficient metric between prediction and ground truth.
def dice_coef_metric(pred, label):
    intersection = 2.0 * (pred * label).sum()
    union = pred.sum() + label.sum()
    if pred.sum() == 0 and label.sum() == 0:
        return 1.
    return intersection / union

# Function to calculate the Dice coefficient loss between prediction and ground truth.
def dice_coef_loss(pred, label):
    smooth = 1.0
    intersection = 2.0 * (pred * label).sum() + smooth
    union = pred.sum() + label.sum() + smooth
    return 1 - (intersection / union)

# Function to calculate the combined BCE (Binary Cross Entropy) and Dice loss.
def bce_dice_loss(pred, label):
    dice_loss = dice_coef_loss(pred, label)
    bce_loss = nn.BCELoss()(pred, label)
    return dice_loss + bce_loss

def iou_metric(pred, label):
    intersection = (pred * label).sum()
    union = pred.sum() + label.sum() - intersection
    if union == 0:
        return 1.0
    else:
        return intersection / union
```

```python
# Function to perform the training loop for the model.
def train_loop(model, loader, loss_func):
    model.train()
    train_losses = []
    train_dices = []
    train_ious = []

    for i, (image, mask) in enumerate(loader):
        image = image.to(device)
        mask = mask.to(device)
        outputs = model(image)

        # Convert outputs to numpy array for post-processing
        out_cut = np.copy(outputs.data.cpu().numpy())
        out_cut[np.nonzero(out_cut < 0.5)] = 0.0
        out_cut[np.nonzero(out_cut >= 0.5)] = 1.0

        dice = dice_coef_metric(out_cut, mask.data.cpu().numpy())
        iou = iou_metric(out_cut, mask.data.cpu().numpy())

        loss = loss_func(outputs, mask)

        train_losses.append(loss.item())
        train_dices.append(dice)
        train_ious.append(iou)

        loss.backward()
        optimizer.step()
        optimizer.zero_grad()

    return train_dices, train_ious, train_losses
```

```python
# Function to perform evaluation loop for the model.
def eval_loop(model, loader, loss_func, training=True):
    model.eval()
    val_loss = 0
    val_dice = 0
    val_iou = 0
    with torch.no_grad():
        for step, (image, mask) in enumerate(loader):
            image = image.to(device)
            mask = mask.to(device)

            outputs = model(image)
            loss = loss_func(outputs, mask)

            # Convert outputs to numpy array for post-processing
            out_cut = np.copy(outputs.data.cpu().numpy())
            out_cut[np.nonzero(out_cut < 0.5)] = 0.0
            out_cut[np.nonzero(out_cut >= 0.5)] = 1.0

            dice = dice_coef_metric(out_cut, mask.data.cpu().numpy())
            iou = iou_metric(out_cut, mask.data.cpu().numpy())

            val_loss += loss
            val_dice += dice
            val_iou += iou

        val_mean_dice = val_dice / step
        val_mean_iou = val_iou / step
        val_mean_loss = val_loss / step

        if training:
            scheduler.step(val_mean_dice)

    return val_mean_dice, val_mean_iou, val_mean_loss
```

```python
# Function to train the model and evaluate on validation data across epochs.
def train_model(train_loader, val_loader, loss_func, optimizer, scheduler, num_epochs):
    train_loss_history = []
    train_dice_history = []
    train_iou_history = []
    val_loss_history = []
    val_dice_history = []
    val_iou_history = []

    for epoch in range(num_epochs):
        train_dices, train_ious, train_losses = train_loop(model, train_loader, loss_func)

        train_mean_dice = np.array(train_dices).mean()
        train_mean_iou = np.array(train_ious).mean()
        train_mean_loss = np.array(train_losses).mean()

        val_mean_dice, val_mean_iou, val_mean_loss = eval_loop(model, val_loader, loss_func)

        train_loss_history.append(train_mean_loss)
        train_dice_history.append(train_mean_dice)
        train_iou_history.append(train_mean_iou)
        val_loss_history.append(val_mean_loss)
        val_dice_history.append(val_mean_dice)
        val_iou_history.append(val_mean_iou)

        print(f"Epoch: {epoch+1}/{num_epochs} | Train Loss: {train_mean_loss:.3f}, Val Loss: {val_mean_loss:.3f}, Train DICE: {train_mean_dice:.3f}, Val DICE: {val_mean_dice:.3f}, Train IOU: {train_mean_iou:.3f}, Val IOU: {val_mean_iou:.3f}")

    return train_loss_history, train_dice_history, train_iou_history, val_loss_history, val_dice_history, val_iou_history
```

```python
# Define the optimizer with Adam optimizer and initial learning rate of 1e-3
optimizer = torch.optim.Adam(model.parameters(), lr=1e-3)

# Define the learning rate scheduler with ReduceLROnPlateau, monitoring 'max' validation metric, and patience of 3 epochs
scheduler = torch.optim.lr_scheduler.ReduceLROnPlateau(optimizer, 'max', patience=3)

# Number of epochs for training
num_epochs = 100
```

```python
from PIL import Image
import torch
import numpy as np
import requests
import torchvision.transforms as tt

def preprocess_image(image_path, target_size=(128, 128)):
    # If the input is a URL, download the image
    if image_path.startswith('http'):
        response = requests.get(image_path)
        img = Image.open(response.content)
    else:
        # If the input is a local file path
        img = Image.open(image_path)

    # Convert grayscale images to RGB if necessary
    if img.mode != 'RGB':
        img = img.convert('RGB')

    # Resize the image to 128x128 (same as during training)
    img = img.resize(target_size)

    # Convert the image to numpy array
    img = np.array(img).astype(np.float32) / 255.0  # Normalize to [0, 1]

    # Change the shape to (C, H, W)
    img = np.transpose(img, (2, 0, 1))  # Convert to (C, H, W)

    # Normalize the image as done in training (e.g., using ImageNet mean and std)
    normalize = tt.Normalize(mean=(0.485, 0.456, 0.406), std=(0.229, 0.224, 0.225))
    img_tensor = torch.tensor(img).unsqueeze(0)  # Add batch dimension
    img_tensor = normalize(img_tensor)

    return img_tensor
```

```python
import gradio as gr
import torch
import numpy as np
from PIL import Image
from lime import lime_image
from skimage.segmentation import mark_boundaries
import torchvision.transforms as T
import torch.nn as nn
import torch.nn.functional as F # Import F for padding

# --- U-Net Components (Copied from earlier in the notebook) ---

class DoubleConv(nn.Module):
    """(convolution => [BN] => ReLU) * 2"""
    def __init__(self, in_channels, out_channels, mid_channels=None):
        super().__init__()
        if not mid_channels:
            mid_channels = out_channels
        self.double_conv = nn.Sequential(
            nn.Conv2d(in_channels, mid_channels, kernel_size=3, padding=1),
            nn.BatchNorm2d(mid_channels),
            nn.ReLU(inplace=True),
            nn.Conv2d(mid_channels, out_channels, kernel_size=3, padding=1),
            nn.BatchNorm2d(out_channels),
            nn.ReLU(inplace=True))
    def forward(self, x):
        return self.double_conv(x)


class Down(nn.Module):
    """Downscaling with maxpool then double conv"""
    def __init__(self, in_channels, out_channels):
        super().__init__()
        self.maxpool_conv = nn.Sequential(
            nn.MaxPool2d(2),
            DoubleConv(in_channels, out_channels))
    def forward(self, x):
        return self.maxpool_conv(x)


class Up(nn.Module):
    """Upscaling then double conv"""
    def __init__(self, in_channels, out_channels, bilinear=True):
        super().__init__()
        if bilinear:
            self.up = nn.Upsample(scale_factor=2, mode='bilinear', align_corners=True)
            self.conv = DoubleConv(in_channels, out_channels, in_channels//2)
        else:
            self.up = nn.ConvTranspose2d(in_channels, in_channels//2, kernel_size=2, stride=2)
            self.conv = DoubleConv(in_channels, out_channels)
    def forward(self, x1, x2):
        x1 = self.up(x1)
        # input is CHW
        diffY = x2.size()[2] - x1.size()[2]
        diffX = x2.size()[3] - x1.size()[3]

        x1 = F.pad(x1, [diffX // 2, diffX - diffX // 2,
                        diffY // 2, diffY - diffY // 2])
        x = torch.cat([x2, x1], dim=1)
        return self.conv(x)

class OutConv(nn.Module):
    def __init__(self, in_channels, out_channels):
        super(OutConv, self).__init__()
        self.conv = nn.Sequential(
            nn.Conv2d(in_channels, out_channels, kernel_size=1),
            nn.Sigmoid()) # Sigmoid is important here for binary segmentation output
    def forward(self, x):
        return self.conv(x)

# --- U-Net Definition (Matching the trained model) ---
class UNet(nn.Module):
    def __init__(self, n_channels, n_classes, bilinear=True):
        super(UNet, self).__init__()
        self.n_channels = n_channels
        self.n_classes = n_classes
        self.bilinear = bilinear

        # Encoder (downsampling path)
        self.inc = DoubleConv(n_channels, 64)
        self.down1 = Down(64, 128)
        self.down2 = Down(128, 256)
        self.down3 = Down(256, 512)
        factor = 2 if bilinear else 1
        self.down4 = Down(512, 1024//factor)

        # Decoder (upsampling path)
        self.up1 = Up(1024, 512//factor, bilinear)
        self.up2 = Up(512, 256//factor, bilinear)
        self.up3 = Up(256, 128//factor, bilinear)
        self.up4 = Up(128, 64, bilinear)

        # Output layer
        self.outc = OutConv(64, n_classes)

    def forward(self, x):
        # Encoder path
        x1 = self.inc(x)
        x2 = self.down1(x1)
        x3 = self.down2(x2)
        x4 = self.down3(x3)
        x5 = self.down4(x4)

        # Decoder path with skip connections
        x = self.up1(x5, x4)
        x = self.up2(x, x3)
        x = self.up3(x, x2)
        x = self.up4(x, x1)

        # Final output
        logits = self.outc(x)
        return logits


# --- Load Model ---
# Instantiate the model with the correct parameters (input channels=3, output channels=1)
unet_model = UNet(n_channels=3, n_classes=1)
# Load the saved state dictionary
unet_model.load_state_dict(torch.load("brain-mri-unet.pth", map_location="cpu"))
# Set the model to evaluation mode
unet_model.eval()

# --- Preprocessing ---
# Use the same normalization and resizing as used in the training DataLoader
def preprocess(img: Image.Image):
    img = img.resize((128, 128)).convert("RGB")
    tensor = T.ToTensor()(img)
    # Use the same normalization values as in the BrainDataset
    normalize = T.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
    tensor = normalize(tensor).unsqueeze(0) # Add batch dimension
    return tensor

# --- Segmentation ---
def segment_image(image_np):
    # Convert numpy input image to PIL Image for preprocessing
    image_pil = Image.fromarray(image_np.astype(np.uint8))
    tensor = preprocess(image_pil)

    # Move the tensor to the appropriate device if your model is on GPU
    # In this Gradio app, we're loading the model to CPU, so no device change needed
    # tensor = tensor.to(next(unet_model.parameters()).device) # Uncomment if using GPU

    with torch.no_grad():
        pred = unet_model(tensor)

    # Post-process the output
    # The OutConv already applies Sigmoid, so no need to apply it again here
    # pred = torch.sigmoid(pred) # Remove this line if OutConv includes Sigmoid
    pred_mask = (pred.squeeze().cpu().numpy() > 0.5).astype(np.uint8) * 255 # Move prediction to CPU before numpy conversion

    return Image.fromarray(pred_mask)

# --- LIME Explanation ---
def lime_explanation(image_np):
    # LIME prediction function needs to accept a batch of numpy images
    # and return a batch of prediction probabilities for the target class
    def batch_predict(images):
        batch = []
        for img in images:
            # Convert numpy image to PIL, preprocess, and add to batch
            img_pil = Image.fromarray(img.astype(np.uint8))
            processed_img_tensor = preprocess(img_pil).squeeze(0) # Remove batch dim for stacking
            batch.append(processed_img_tensor)

        batch_tensor = torch.stack(batch)

        # Move the tensor to the appropriate device if your model is on GPU
        # batch_tensor = batch_tensor.to(next(unet_model.parameters()).device) # Uncomment if using GPU

        with torch.no_grad():
            logits = unet_model(batch_tensor)
            # Get probability for the positive class (assuming it's the output channel 0)
            probs = logits.squeeze().cpu().numpy() # Assuming logits are directly probabilities after Sigmoid in OutConv
            # If OutConv output is not [0, 1], apply sigmoid:
            # probs = torch.sigmoid(logits).squeeze().cpu().numpy()

        # Reshape to (num_samples, num_classes). For binary segmentation, num_classes = 1.
        # LIME expects (num_samples, flattened_output_shape).
        # Since we are explaining a single pixel or area, the output is spatial.
        # We need to flatten the spatial output for LIME's classifier explanation.
        # However, for image explanation, LIME needs a score per superpixel.
        # A common approach for segmentation is to treat the probability map for the target class as the "score".
        # We return the flattened probability map.
        return probs.reshape((len(images), -1)) # Flatten the spatial dimensions


    explainer = lime_image.LimeImageExplainer()
    image = Image.fromarray(image_np.astype(np.uint8))
    # Resize the input image to the model's expected input size for LIME
    img_resized = np.array(image.resize((128, 128)).convert("RGB"))

    # LIME explanation expects a classifier prediction function that returns
    # probabilities for each class. For segmentation, we often explain the
    # prediction of the tumor class (assuming it's class 1).
    # The `batch_predict` function above is adapted to return the flattened
    # probability map for the tumor class. LIME then uses this to explain
    # which superpixels contribute to a high score (probability) in the output.

    # The `top_labels` argument refers to the index of the class we want to explain.
    # Since our U-Net has 1 output channel (for binary segmentation),
    # we are essentially explaining the probability of the tumor class.
    # LIME treats the flattened output as a single 'class' or score for explanation.
    explanation = explainer.explain_instance(
        img_resized,
        batch_predict,
        # For a flattened segmentation output, LIME doesn't directly map to a single class index.
        # We are explaining the presence/absence of tumor.
        # A common workaround is to explain based on an average score or a specific pixel.
        # Here, we'll rely on LIME's internal handling for image explanation.
        # `top_labels=1` might not be strictly correct for a flattened output,
        # but LIME's image explainer can still provide a spatial explanation.
        top_labels=1, # This argument might need adjustment depending on how LIME interprets the flattened output
        hide_color=0,
        num_samples=100
    )
    # The mask from get_image_and_mask corresponds to regions that contribute
    # positively or negatively to the explained 'class' (here, the flattened tumor probability).
    # For positive_only=True, it highlights regions contributing positively.
    temp, mask = explanation.get_image_and_mask(
        explanation.top_labels[0], # Use the index of the top predicted 'class' from the flattened output
        positive_only=True,
        hide_rest=False, # Show the rest of the image faintly
        num_features=10 # Number of superpixels to highlight
    )
    # mark_boundaries overlays the mask on the original image (temp)
    lime_result = mark_boundaries(temp, mask)
    return Image.fromarray((lime_result * 255).astype(np.uint8))




# --- Custom Styling ---
custom_css = """
.gradio-container {

    background-image: url(' https://wallpapers.com/images/hd/white-color-background-ghw6e685ri75chj4.jpg ');
    background-size: cover;
    font-family: 'Segoe UI', sans-serif;
    padding: 30px;
}
h1 {
    text-align: center;
    font-size: 20px;
    color: #3b0066;
    margin-bottom: 0.2em;
}
p {
    text-align: center;
    font-size: 1.1em;
    color: #333;
}
"""
# --- Gradio UI ---
with gr.Blocks(css=custom_css) as demo:
    gr.Markdown("<h1>ASTROCYTOMA XAI ANALYZER</h1>")
    gr.Markdown("<p>Upload an MRI image to get tumor segmentation and LIME explanation</p>")

    with gr.Row():
        # ------------------- CHANGED HERE  -------------------
        # Set a larger display size (256256) for the input image
        input_img = gr.Image(type="numpy", label="Upload MRI Image",
                              interactive=True,
                              height=256, width=256)
        # -------------------------------------------------------

    analyze_btn = gr.Button("Analyze")

    with gr.Row():
        # ------------------- CHANGED HERE  -------------------
        # Also enlarge both output panels to 256256
        output_segment = gr.Image(label="Segmented Output", height=412, width=412)
        output_lime    = gr.Image(label="LIME Explanation", height=412, width=412)
        # -------------------------------------------------------

    analyze_btn.click(
        fn=lambda img: (segment_image(img), lime_explanation(img)),
        inputs=input_img,
        outputs=[output_segment, output_lime]
    )

demo.launch()
# --- Gradio UI ---
"""with gr.Blocks(css=custom_css) as demo:
    gr.Markdown("<h1>ASTROCYTOMA XAI ANALYZER</h1>")
    gr.Markdown("<p>Upload an MRI image to get tumor segmentation and LIME explanation</p>")

    with gr.Row():
        # Use type="numpy" for easier handling with OpenCV and LIME
        input_img = gr.Image(type="numpy", label="Upload MRI Image")

    analyze_btn = gr.Button("Analyze")

    with gr.Row():
        output_segment = gr.Image(label="Segmented Output")
        output_lime = gr.Image(label="LIME Explanation")

    analyze_btn.click(
        # Pass the numpy image directly to the functions
        fn=lambda img: (segment_image(img), lime_explanation(img)),
        inputs=input_img,
        outputs=[output_segment, output_lime]
    )

demo.launch()"""
```

```python
import gradio as gr
import torch
import numpy as np
from PIL import Image
from lime import lime_image
from skimage.segmentation import mark_boundaries
import torchvision.transforms as T
import torch.nn as nn
import torch.nn.functional as F  # Import F for padding

# --- U-Net Components (copied from earlier) ---

class DoubleConv(nn.Module):
    """(convolution => [BN] => ReLU) * 2"""
    def __init__(self, in_channels, out_channels, mid_channels=None):
        super().__init__()
        if not mid_channels:
            mid_channels = out_channels
        self.double_conv = nn.Sequential(
            nn.Conv2d(in_channels, mid_channels, kernel_size=3, padding=1),
            nn.BatchNorm2d(mid_channels),
            nn.ReLU(inplace=True),
            nn.Conv2d(mid_channels, out_channels, kernel_size=3, padding=1),
            nn.BatchNorm2d(out_channels),
            nn.ReLU(inplace=True))
    def forward(self, x):
        return self.double_conv(x)


class Down(nn.Module):
    """Downscaling with maxpool then double conv"""
    def __init__(self, in_channels, out_channels):
        super().__init__()
        self.maxpool_conv = nn.Sequential(
            nn.MaxPool2d(2),
            DoubleConv(in_channels, out_channels))
    def forward(self, x):
        return self.maxpool_conv(x)


class Up(nn.Module):
    """Upscaling then double conv"""
    def __init__(self, in_channels, out_channels, bilinear=True):
        super().__init__()
        if bilinear:
            self.up = nn.Upsample(scale_factor=2, mode='bilinear', align_corners=True)
            self.conv = DoubleConv(in_channels, out_channels, in_channels // 2)
        else:
            self.up = nn.ConvTranspose2d(in_channels, in_channels // 2, kernel_size=2, stride=2)
            self.conv = DoubleConv(in_channels, out_channels)
    def forward(self, x1, x2):
        x1 = self.up(x1)
        # input is CHW
        diffY = x2.size()[2] - x1.size()[2]
        diffX = x2.size()[3] - x1.size()[3]

        x1 = F.pad(x1, [diffX // 2, diffX - diffX // 2,
                        diffY // 2, diffY - diffY // 2])
        x = torch.cat([x2, x1], dim=1)
        return self.conv(x)


class OutConv(nn.Module):
    def __init__(self, in_channels, out_channels):
        super(OutConv, self).__init__()
        self.conv = nn.Sequential(
            nn.Conv2d(in_channels, out_channels, kernel_size=1),
            nn.Sigmoid())  # Sigmoid for binary segmentation
    def forward(self, x):
        return self.conv(x)


class UNet(nn.Module):
    def __init__(self, n_channels, n_classes, bilinear=True):
        super(UNet, self).__init__()
        self.n_channels = n_channels
        self.n_classes = n_classes
        self.bilinear = bilinear

        # Encoder
        self.inc = DoubleConv(n_channels, 64)
        self.down1 = Down(64, 128)
        self.down2 = Down(128, 256)
        self.down3 = Down(256, 512)
        factor = 2 if bilinear else 1
        self.down4 = Down(512, 1024 // factor)

        # Decoder
        self.up1 = Up(1024, 512 // factor, bilinear)
        self.up2 = Up(512, 256 // factor, bilinear)
        self.up3 = Up(256, 128 // factor, bilinear)
        self.up4 = Up(128, 64, bilinear)

        # Output
        self.outc = OutConv(64, n_classes)

    def forward(self, x):
        x1 = self.inc(x)
        x2 = self.down1(x1)
        x3 = self.down2(x2)
        x4 = self.down3(x3)
        x5 = self.down4(x4)

        x = self.up1(x5, x4)
        x = self.up2(x, x3)
        x = self.up3(x, x2)
        x = self.up4(x, x1)

        logits = self.outc(x)
        return logits


# --- Load Model ---
unet_model = UNet(n_channels=3, n_classes=1)
unet_model.load_state_dict(torch.load("brain-mri-unet.pth", map_location="cpu"))
unet_model.eval()

# --- Preprocessing ---
def preprocess(img: Image.Image):
    img = img.resize((128, 128)).convert("RGB")
    tensor = T.ToTensor()(img)
    normalize = T.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
    tensor = normalize(tensor).unsqueeze(0)
    return tensor

# --- Segmentation ---
def segment_image(image_np):
    """
    Returns a PIL Image of the tumor mask in dark green (RGB) with a thick outline.
    """
    image_pil = Image.fromarray(image_np.astype(np.uint8))
    tensor = preprocess(image_pil)
    with torch.no_grad():
        pred = unet_model(tensor)

    # ------------------- CHANGED HERE  -------------------
    # Create a binary mask (0 or 1) from the sigmoided output:
    pred_mask_bin = (pred.squeeze().cpu().numpy() > 0.5).astype(np.uint8)

    # Convert to an RGB mask of size (128, 128, 3), where tumor pixels are dark green (0,128,0)
    h, w = pred_mask_bin.shape
    green_mask = np.zeros((h, w, 3), dtype=np.uint8)
    green_mask[pred_mask_bin == 1] = [0, 128, 0]       # Darkgreen for tumor
    # Keep background as black [0,0,0]
    # Now draw a thick boundary around the green region:
    # We can use mark_boundaries with mode='thick'
    green_outline = mark_boundaries(green_mask / 255.0, pred_mask_bin, color=(0, 100/255, 0), mode='thick')
    # Convert from float [0..1] back to uint8 [0..255]
    green_outline_uint8 = (green_outline * 255).astype(np.uint8)

    return Image.fromarray(green_outline_uint8)
    # -------------------------------------------------------


# --- LIME Explanation ---
def lime_explanation(image_np):
    """
    Returns a PIL Image with a LIME overlay in dark green and thick boundaries.
    """
    def batch_predict(images):
        batch = []
        for img in images:
            img_pil = Image.fromarray(img.astype(np.uint8))
            processed_img_tensor = preprocess(img_pil).squeeze(0)
            batch.append(processed_img_tensor)
        batch_tensor = torch.stack(batch)
        with torch.no_grad():
            logits = unet_model(batch_tensor)
            # logits are already after Sigmoid in OutConv
            probs = logits.squeeze().cpu().numpy()
        # Flatten spatial dims so LIME can treat it like one score.
        return probs.reshape((len(images), -1))

    explainer = lime_image.LimeImageExplainer()
    image_pil = Image.fromarray(image_np.astype(np.uint8))
    img_resized = np.array(image_pil.resize((128, 128)).convert("RGB"))

    explanation = explainer.explain_instance(
        img_resized,
        batch_predict,
        top_labels=1,
        hide_color=0,
        num_samples=100
    )

    # ------------------- CHANGED HERE  -------------------
    # Get the top label (flattenedoutputindex) and fetch its mask:
    top_label = explanation.top_labels[0]

    # positive_only=True  only highlight superpixels that contribute positively
    temp, mask = explanation.get_image_and_mask(
        top_label,
        positive_only=True,
        hide_rest=False,
        num_features=10
    )
    # temp is the image with superpixels colored; mask is a labelmap of those superpixels

    # Use mark_boundaries with dark green color and thick mode
    lime_overlay = mark_boundaries(temp, mask, color=(0, 100/255, 0), mode='thick')
    lime_overlay_uint8 = (lime_overlay * 255).astype(np.uint8)
    return Image.fromarray(lime_overlay_uint8)
    # -------------------------------------------------------


# --- Custom CSS (unchanged except for optional finetuning) ---
custom_css = """
.gradio-container {
    background-image: url('https://png.pngtree.com/background/20210717/original/pngtree-colored-watercolor-splash-border-background-picture-image_1432675.jpg');
    background-size: cover;
    font-family: 'Segoe UI', sans-serif;
    padding: 30px;
}
h1 {
    text-align: center;
    font-size: 2.7em;
    color: #3b0066;
    margin-bottom: 0.2em;
}
p {
    text-align: center;
    font-size: 1.1em;
    color: #333;
}
"""


# --- Gradio UI ---
with gr.Blocks(css=custom_css) as demo:
    gr.Markdown("<h1>ASTROCYTOMA XAI ANALYZER</h1>")
    gr.Markdown("<p>Upload an MRI image to get tumor segmentation and LIME explanation</p>")

    with gr.Row():
        # ------------------- CHANGED HERE  -------------------
        # Set a larger display size (256256) for the input image
        input_img = gr.Image(type="numpy", label="Upload MRI Image",
                              interactive=True,
                              height=256, width=256)
        # -------------------------------------------------------

    analyze_btn = gr.Button("Analyze")

    with gr.Row():
        # ------------------- CHANGED HERE  -------------------
        # Also enlarge both output panels to 256256
        output_segment = gr.Image(label="Segmented Output", height=256, width=256)
        output_lime    = gr.Image(label="LIME Explanation", height=256, width=256)
        # -------------------------------------------------------

    analyze_btn.click(
        fn=lambda img: (segment_image(img), lime_explanation(img)),
        inputs=input_img,
        outputs=[output_segment, output_lime]
    )

demo.launch()
```
